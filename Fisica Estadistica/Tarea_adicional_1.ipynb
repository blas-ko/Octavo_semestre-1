## 2) Usando la información anterior calcule y grafique la distribución de la magnitud de las velocidad $v$ que toma el átomo.

Vamos a crear una función que cuente cuantas veces se tienen las velocidades en ciertos intervalos.

```julia
function histograma(X::Array,divisiones::Int)
    Xmin=minimum(X)
    Xmax=maximum(X)
    intervalo=Xmax-Xmin
    INTERVALOS=[Xmin:intervalo/divisiones:Xmax]
    contadora=zeros(Int,divisiones)
    
    for i in 1:divisiones-1
        contadora[i]=count(x-> INTERVALOS[i]<=x<INTERVALOS[i+1],X)
    end
    
    contadora[end]=count(x-> INTERVALOS[end-1]<=x<=INTERVALOS[end],X)
    Plotting_itvls=zeros(divisiones)
    
    for i in 1:divisiones
        Plotting_itvls[i]=(INTERVALOS[i+1]+INTERVALOS[i])/2
    end
    
    return contadora,Plotting_itvls
end
```

```julia
function normal_histo(X::Array,divisiones::Int)
    H,INTERVALOS=histograma(X,divisiones)
    TOTAL=sum(H)
    Hnorm=H/TOTAL
    return Hnorm,INTERVALOS
end
```

```julia
@time W,V_s=normal_histo(V,100);
```

```julia
figure(figsize=(10,5))
xlabel("Velocidad (m/s)")
ylabel("Probabilidad")
title("Distribución de probabilidad de las velocidades para el átomo.")
plot(V_s,W);
```

Esta distribución de velocidades se obtuvo entonces haciendo una partición en 100 divisiones de todos los valores de las velocidades, y de allí contando cuantas velocidades se obtuvieron dentro de cada intervalo. Al final se normalizó la distribución obtenida, dividiendo cada número del arreglo obtenido por el número total de mediciones.

## 3) De los datos originales $(v_x, v_y, v_z)$ calcule el promedio de la magnitud de la velocidad $<v>$ y de la fluctuación (varianza) $\sqrt{<v^2>-<v>^2}$.

Vamos a calcular los promedios de cada uno de los valores:

```julia
Promedio_v_cuadrado_3=mean(vel2);
```

## 5)

A "ojo", la velocidad más probable parecería ser aproximadamente $700 \frac{m}{s}$. Vamos a hacer una función rápida para ver qué velocidad corresponde al valor máximo de $\mathcal{W}(v)$:

```julia
function vel_maxima(Vels::Array, W::Array)
    Wmax=maximum(W)
    FIND=zeros(W)
    vel=0
    for i in 1:length(W)
        if W[i]==Wmax
            FIND[i]=1
        end
        vel+=FIND[i]*Vels[i]
    end
    vel
end;
```

```julia
V_max=vel_maxima(V_s,W)
```

Podemos ver como a "super ojo", la velocidad más probable corresponde a $678.09 \frac{m}{s}$, teniendo entonces que difiere de la velocidad promedio. Esto es lógico ya que la distribución no parece ser simétrica con respecto al máximo, y está como "estirada" hacia la derecha, aumentando el promedio de velocidades.

Finalmente, vamos a hacer una gráfica que muestre la distribución de las velocidades, así como la velocidad promedio, la que tiene máxima probabilidad y la fluctuación:

```julia
figure(figsize=(10,5))
xlabel("Velocidad (m/s)")
ylabel("Probabilidad")
title("Distribución de probabilidad de las velocidades para el átomo.")
plot(V_s,W)
plot([Promedio_v_3,Promedio_v_3],[0,0.03], label="Velocidad promedio")
plot([V_max,V_max],[0,0.03], label="Velocidad más probable")
plot([Promedio_v_3-Fluctuacion_3,Promedio_v_3,Promedio_v_3+Fluctuacion_3],[0.001,0.001,0.001],"o-", label="Fluctuación")
legend(loc=0);
```
