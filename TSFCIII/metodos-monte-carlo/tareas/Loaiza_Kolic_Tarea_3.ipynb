```julia
function next_walker(t::Int,i::Int,der,cen,izq,P0,largo=(length(P0)-1)/2,Itvl0=[-largo:largo])
    
    if der+cen+izq!=1
        error("La probabilidad total es distinta a 1")
    end
    
    if t==0
        if i in Itvl0
            return P0[largo+i+1]
        else
            return 0
        end
    else
        
        return der*next_walker(t-1,i-1,der,cen,izq,P0,largo,Itvl0)+izq*next_walker(t-1,i+1,der,cen,izq,P0,largo,Itvl0)+cen*next_walker(t-1,i,der,cen,izq,P0,largo,Itvl0)
    end
end;
```

```julia
function walker_array_1D(t::Int,der,cen,izq,P0)
    
    P=zeros(2t+1)
    
    Itvl=[-t:t]
    
    for i in Itvl
        P[i+t+1]=next_walker(t,i,der,cen,izq,P0)
    end
    
    return P,Itvl
end
```

## 2)

```julia
function graficadora(t::Int,der,cen,izq,P0=[1])
    Resultados=Array[]
    Intervalos=Array[]

    for i in 0:t
        R,I=walker_array_1D(i,der,cen,izq,P0)
        push!(Resultados,R)
        push!(Intervalos,I)
    end
    
    Resultados,Intervalos
    
    
    figure(figsize=(10,5))
    
    for i in t:-1:0
        plot(Intervalos[i+1]/sqrt(i+1),sqrt(i+1)*Resultados[i+1],label="t=$i","x-",linewidth=3, markersize=10)  
    end
    
    xlabel(L"$\frac{x}{\sqrt{t}}$")
    ylabel(L"$\sqrt{t}*P_t(x)$")  
    legend(loc=0,fontsize=10)
    title("Probabilidades: izquierda=$(round(izq,3)), centro=$(round(cen,3)), derecha=$(round(der,3)). \n Inicio en $P0 \n")
end;
```

```julia
function next_walker_zero(L::Int,t::Int,x0::Int,i::Int)
    if t==0
        if i==x0
            return 1
        else
            return 0
        end
    else
        
        if i<0 || i>L #si no está en el área de interés tiene que ser 0 la probabilidad
            return 0
        end
        
        
        if 1<i<L #si está "en el centro" es la forma recursiva clásica la probabilidad
            return (1/3)*next_walker_zero(L,t-1,x0,i-1)+(1/3)*next_walker_zero(L,t-1,x0,i+1)+(1/3)*next_walker_zero(L,t-1,x0,i)
        end
        
        if i==1 && L != 1 #si está en 1, como 0 se "come la probabilidad" hay que quitarle esa parte
            return (1/3)*next_walker_zero(L,t-1,x0,2)+(1/3)*next_walker_zero(L,t-1,x0,1)
        end
        
        if i==1 && L==1
            return (1/3)*next_walker_zero(L,t-1,x0,2)+(1/3)*next_walker_zero(L,t-1,x0,1)+(1/3)*next_walker_zero(L,t-1,x0,1)
        end
        
        if i==0 #si está en 0 sólo pudo haber saltado del 1 ya que 0 se come la probabilidad y no hay en <0
            return (1/3)*next_walker_zero(L,t-1,x0,i+1)
        end
        
        if i==L #si salta a la derecha en L rebota, por lo cual no viene de la izquierda pero si viene "dos veces" de L
            return (1/3)*next_walker_zero(L,t-1,x0,i-1)+(1/3)*next_walker_zero(L,t-1,x0,i)+(1/3)*next_walker_zero(L,t-1,x0,i)
        end
    end
end;
```

```julia
function walker_zero_array_1D(L::Int,x0::Int,t::Int)
    
    P=zeros(2t+1)
    
    Itvl=[-t:t]
    
    for i in Itvl
        P[i+t+1]=next_walker_zero(L,t,x0,i)
    end
    
    return P[t+1]
end;
```

```julia
function graficadora_zero(L::Int,t::Int,x0::Int)
    Zeros=Float64[]

    for i in 0:t
        P0=walker_zero_array_1D(L,x0,i)
        push!(Zeros,P0)
    end
    
    plot([0:t],Zeros,label="x0=$x0, L=$L")
    xlabel("t pasos")
    ylabel(L"$P_t(0)$")  
    legend(loc=0,fontsize=10)
    #title("Probabilidad en función de los pasos de llegar al cero \n posición inicial en x0=$x0 y pared en L=$L \n")
end;
```

Debemos de tener que:
$$\sum_{i=0}^{\infty}P_i(0)=1$$
Sea que en tiempo infinito, todos los caminantes tocaron al 0.
