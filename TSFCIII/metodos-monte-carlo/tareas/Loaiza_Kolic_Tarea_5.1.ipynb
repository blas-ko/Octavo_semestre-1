# Cadenas de Markov 

Un caminante aleatorio es un tipo particular de una *cadena de Markov*.

Una cadena de Markov es, a su vez, un tipo de 
*proceso estocástico*, es decir una evolución en 
el tiempo que no es determinista, sino donde hay 
una probabilidad dada, el sistema lleva a cabo 
una *transición* hacia un estado nuevo. Las cadenas de Markov son de gran importancia hoy día en las ciencias.

La particularidad de las cadenas de Markov es que satisfacen la *propiedad de Markov*: el estado nuevo depende *sólo de un estado anterior*. Se suele decir que el proceso *no tiene memoria*.

Por ejemplo, una caminata aleatoria satisface esto, ya que la posición al tiempo $t+1$ depende sólo de la posición al tiempo $t$ (no de posiciones a tiempos anteriores).  

Sea $X_n$ el estado del proceso al tiempo $n$. Nos interesaremos por el momento en procesos de Markov tal que
$P(X_{n+1} = j \, | \, X_n = i) = p_{ij}$, independiente del tiempo. La notación aquí es de una *probabilidad condicional*, es decir, la probabilidad de que el sistema está en el estado $j$ al tiempo $n+1$, *dado que* estaba en el estado $i$ al tiempo $n$.

## 1)

### $i)$

Tenemos que la probabilidad de pasar a uno de los estados "al lado" es $p, \ q \text{ ó }r$, y si se está en uno de los bordes es $p+r$ ó $q+r$, dependiendo de que lado esté. La matriz de transición se ve entonces como ($p$ es la probabilidad de saltar a la derecha, $q$ a la izquierda y $r$ de quedarse en el mismo lugar):
<p>
$$\begin{bmatrix}
q+r & q & 0 & 0 & 0 & 0 \\
p & r & q & 0 & 0 & 0 \\
0 & p & r & q & 0 & 0 \\
0 & 0 & p & r & q & 0 \\
0 & 0 & 0 & p & r & q \\
0 & 0 & 0 & p & p & r+p \\
\end{bmatrix}$$

### $ii)$

Cualquier matriz de transición (de $N\times N$) debe de cumplir la propiedad siguiente:
$$\sum_{j=1}^{N}p_{ij}=1$$

Y claramente también debe de ser una matriz cuadrada (ya que sino no nos está dando toda la información de las posibilidades y no sirve para operar).

## 2)

### $i)$

Sabemos que la matriz debe de ser de $4\times 4$, que debe de satisfacer $\sum_{i=1}^4 p_{ij}=1$ y como todas las probabilidades son iguales entonces $p_{ij}=p_{kl}$ para todas $i,j,k,l\in \{ 1,2,3,4 \} $. Como son iguales, tenemos entonces que:
$$\sum_{i=1}^{4}p_{11}=4*p_{11}=1,$$
Llegando finalmente a que:
$$p_{ij}=\frac{1}{4} \ \forall i,j\in \{ 1,2,3,4 \}$$

De aquí llegamos a que la matriz debe de estar descrita como:
$$\begin{bmatrix}
0 & 1/3 & 1/3 & 1/3 \\
1/3 & 0 & 1/3 & 1/3 \\
1/3 & 1/3 & 0 & 1/3 \\
1/3 & 1/3 & 1/3 & 0 \\
\end{bmatrix}$$

### $ii)$

Las propiedades que nos podrían interesar es cuál va a ser el estado no dentro de un paso sino dentro de varios. Para hacer esto necesitaríamos multiplicar la matriz por ella misma $M$ veces, y se obtendría entonces otra matriz de $N\times N$ en la cual:
$$ p_{ij}=P(X_{M+1} = j \, | \, X_n = i)$$

También nos podría interesar que la matriz sea invertible (para saber "de donde viene" un estado de probabilidad) y sus eigenvalores.

### $iii)$

Vamos a escribir una función que acepta como entrada una matriz de transición $M$, un estado inicial $\psi$ (el cual es un vector dando la distribución de probabilidades inicial en el sistema) y un número de realizaciones $n$, y se obtiene un estado (probabilista) al cabo de $n$ realizaciones:

```julia
function histograma(X::Array,divisiones::Int)
    #esta función toma como entrada un arreglo y un número de divisiones, y a la salida escupe el número de elementos que
    #había en cacda uno de los intervalos (las divisiones son homogéneas)
    #además escupe un arreglo que tiene un elemento en la mitad de cada división para poder graficar
    
    Xmin=minimum(X)
    Xmax=maximum(X)
    intervalo=Xmax-Xmin
    INTERVALOS=[Xmin:intervalo/divisiones:Xmax]
    contadora=zeros(Int,divisiones)
    
    for i in 1:divisiones-1
        contadora[i]=count(x-> INTERVALOS[i]<=x<INTERVALOS[i+1],X)
    end
    
    contadora[end]=count(x-> INTERVALOS[end-1]<=x<=INTERVALOS[end],X)
    Plotting_itvls=zeros(divisiones)
    
    for i in 1:divisiones
        Plotting_itvls[i]=(INTERVALOS[i+1]+INTERVALOS[i])/2
    end
    
    return contadora,Plotting_itvls
end

function normal_histo(X::Array,divisiones::Int)
    #hace lo mismo que el histograma pero de forma normalizada
    H,INTERVALOS=histograma(X,divisiones)
    TOTAL=sum(H)
    Hnorm=H/TOTAL
    return Hnorm,INTERVALOS
end;
```

```julia
function step(pos::Int, L::Int)
    new_pos=rand(1:L)
    
    while new_pos==pos
        new_pos=rand(1:L)
    end
    
    return new_pos
end;
```

```julia
function MC_steps(pos::Int, L::Int, N::Int)
    
    S=Int[]
    sizehint(S,N)
    
    push!(S,pos)
    
    for i in 1:N
        push!(S,step(S[i],L))
    end
    
    return S
end;
```

```julia
for i in [10,100,1_000,10_000]
    S=MC_steps(1,4,i)
    Shist,Itvls=normal_histo(S,4)
    plot([1:4],Shist,"o-",label="$i iteraciones")
end
xlabel("Posición")
ylabel("Probabilidad de estar en esa posición \n") 
title("Distribuciones de probabilidad para distinto número de iteraciones \n")
axis([1,4,0.2,0.3])
legend(loc=0, fontsize=10);
```

## 3)

```julia
function arbitrary_step(pos::Int, P, L::Int)
    transition_prob=cumsum( P[pos,:]')
    
    index=0
    rndm=rand()
    
    for i in 1:L
        if transition_prob[i] > rndm
            index=i
            break
        end
    end
    
    return index
    
end;
```

```julia
arbitrary_step(2,P,3)
```

```julia
function MC_arbitrary_steps(pos::Int, L::Int, N::Int, P)
    
    S=zeros(Int,N+1)
    S[1]=pos
    
    for i in 1:N
        S[i+1]=arbitrary_step(S[i],P,L)
    end
    
    return S
end;
```

```julia
P=[1/3 1/3 1/3; 0.5 0.5 0; 0.1 0.9 0]
```

```julia
N=10_000
pos=2
L=3
S=MC_arbitrary_steps(pos,L,N,P)
Shist,Itvls=normal_histo(S,3)
plot([1:3],Shist,"o-")
title("Distribución de probabilidad obtenida al cabo de 10'000 pasos \n ")
xlabel("Posición i")
ylabel("Probabilidad de estar en i");
```

[5] Simula la dinámica de la cadena de Markov de la pregunta 2 desde una condición inicial que es una delta en algún estado por un tiempo $t$. ¿Qué observas cuando $t \to \infty$?

```julia
function realizator(M,ψ,n=1)
    Mn=M
    if n!=1
        for i in 1:n
            Mn*=M
        end
    end
    
    return ψ'*Mn
end;
```

```julia
P=(1/3)*(ones(4,4)-eye(4))
ψ=[1,0,0,0];
```

```julia
for n in 1:2:11
    plot([1:4],(realizator(P,ψ,n))',"o-",label="$n iteraciones")
end
xlabel("Posición i")
ylabel("Probabilidad de estar en i")
title("Distribución exacta de probabilidades para distintos tiempos")
legend(loc=0, fontsize=10);
```

Cuando $t\rightarrow \infty$, la distribución parece aproximarse a una distribución uniforme.
