## 1)

```julia
function caminante(L::Int)
    posicion=1
    pasos=0
    
    while posicion!=0
        pasos+=1
        
        if rand()<0.5
            posicion+=1
            
            if posicion==L+1
                posicion=L
            end
            
        else
            posicion+=-1
        end
    end
    
    return pasos
end
```

## 2)

```julia
function t_promedio(N::Int,L::Int)
    T=0.0
    
    for i in 1:N
        T+=caminante(L)
    end
    
    return T/N
end
```

```julia
@time t_promedio(100_000,10_000);
```

## 3)

```julia
function N_tiempos_promedio(N::Int,L_ini::Int,L_max::Int,L_step::Int)
    [t_promedio(N,L) for L in L_ini:L_step:L_max]
end
```

```julia
@time Tiempos=N_tiempos_promedio(10_000,1,1000,1);
```

```julia
figure(figsize=(10,5))
plot(Tiempos)
xscale("log")
yscale("log")
xlabel("Número de pasos L")
ylabel(L"$<\tau_L>$")
title("Tiempos promedio en función del número de pasos en escala logarítmica");
```

```julia
figure(figsize=(10,5))
plot(Tiempos)
xlabel("Número de pasos L")
ylabel(L"$<\tau_L>$")
title("Tiempos promedio en función del número de pasos en escala normal");
```

Al parecer, $<\tau_L>$ escala linealmente respecto a $L$, teniendo entonces que:

De aquí, claramente podemos decir que, cuando $L \rightarrow \infty$, se tiene que $<\tau_L> \rightarrow \infty$

## 4)

```julia
function tiempos_caminante(N::Int,L::Int)
    T=Int[]
    sizehint(T,N)
    for i in 1:N
        push!(T,caminante(L))
    end
    
    return T
end
```

Esta función va a sacar todo el arreglo de tiempos. De aquí, utilizaremos una "función histogramadora" que, dado un arreglo, regresa sus valores en intervalos listos para hacer un histograma.

```julia
function histograma(X::Array,divisiones::Int)
    #esta función toma como entrada un arreglo y un número de divisiones, y a la salida escupe el número de elementos que
    #había en cacda uno de los intervalos (las divisiones son homogéneas)
    #además escupe un arreglo que tiene un elemento en la mitad de cada división para poder graficar
    
    Xmin=minimum(X)
    Xmax=maximum(X)
    intervalo=Xmax-Xmin
    INTERVALOS=[Xmin:intervalo/divisiones:Xmax]
    contadora=zeros(Int,divisiones)
    
    for i in 1:divisiones-1
        contadora[i]=count(x-> INTERVALOS[i]<=x<INTERVALOS[i+1],X)
    end
    
    contadora[end]=count(x-> INTERVALOS[end-1]<=x<=INTERVALOS[end],X)
    Plotting_itvls=zeros(divisiones)
    
    for i in 1:divisiones
        Plotting_itvls[i]=(INTERVALOS[i+1]+INTERVALOS[i])/2
    end
    
    return contadora,Plotting_itvls
end
```

```julia
function normal_histo(X::Array,divisiones::Int)
    #hace lo mismo que el histograma pero de forma normalizada
    H,INTERVALOS=histograma(X,divisiones)
    TOTAL=sum(H)
    Hnorm=H/TOTAL
    return Hnorm,INTERVALOS
end
```
