En el notebook anterior, vimos cómo muestrear de una distribución de probabilidad $\pi$ dada, usando el algoritmo de Metropolis, que es un ejemplo de un método de "Markov Chain Monte Carlo" ("Monte Carlo usando cadenas de Markov",  MCMC).
Apliquemos esto al modelo de Ising.

$\newcommand{\ss}{\pmb \sigma}
\newcommand{\tt}{\pmb \tau}$

[1] Escribe la probabilidad de aceptación, $\alpha(\ss \to \tt)$, para la distribución de Boltzmann, en términos de la diferencia de energías, $\Delta E := E(\tt) - E(\ss)$.

$$\alpha(\sigma\rightarrow\tau)=\frac{\pi(\tau)}{\pi(\sigma)}=e^{-\beta(E(\tau)-E(\sigma))}.$$

Llegamos entonces a que:
$$\alpha(\sigma\rightarrow\tau)=e^{-\beta\Delta E}$$

[2] En el caso de voltear un solo espín, encuentra una expresión analítica para $\Delta E$, en términos del valor actual del espín que se va a voltear (antes de que se cambie su valor) y los valores de sus espines. 

$$\Delta E=E(\tau)-E(\sigma)$$

Cada una de estas energías es una suma periódica. Como en $E(\tau)$ sólo cambiamos un espín, tenemos que en la resta se van a cancelar todas las componentes de la suma, menos las que están alrededor del espín. Estas, para $E(\sigma)$, son (siendo (i,j) las coordenadas del espín que cambiamos, y $S_{i,j}$ el espín (de $\sigma$) con coordenadas (i,j)):
$$E(\sigma)=\Sigma+S_{i,j}(S_{i-1,j}+S_{i+1,j}+S_{i,j-1}+S_{i,j+1}),$$
(tomando, cuando sea necesario algunos módulos).
Y se tiene que:
$$E(\tau)=\Sigma-S_{i,j}(S_{i-1,j}+S_{i+1,j}+S_{i,j-1}+S_{i,j+1}).$$
Llegamos entonces a que:
$$\Delta E=-2S_{i,j}(S_{i-1,j}+S_{i+1,j}+S_{i,j-1}+S_{i,j+1}).$$

[3] Así, implementa el algoritmo de Metropolis para simular el modelo de Ising a una temperatura $T$ dada. 

Nota que el resultado de la pregunta [2] muestra que, una vez propuesto un espín por voltear, *no* es necesario voltear el espín para calcular la $\Delta E$, y, por lo tanto, se requiere voltearlo *sólo* si el cambio propuesto resulte ser aceptado. Se puede voltear al modificar la misma configuración.

```julia
function config_maker(L::Int)
    
    S=zeros(L,L)
    
    for i in 1:L, j in 1:L
        if rand()<0.5
            S[i,j]=-1
        else
            S[i,j]=1
        end
    end
    
    return S
end;
```

```julia
function spin_choose(L)
    
    i=rand(1:L)
    j=rand(1:L)
    
    return i,j
end;
```

```julia
function spin_flip!(S::Array,i,j,N=length(S),L=int(sqrt(N)))
    
    S[i,j]*=-1
    
    return S    
end;
```

```julia
function periodic_energy(S,N=length(S),L=int(sqrt(N)),J=1)
    
    E=0
    
    for i in 1:L, j in 1:L
        E+=S[i,j]*(S[mod1(i-1,L),j]+S[mod1(i+1,L),j]+S[i,mod1(j-1,L)]+S[i,mod1(j+1,L)])
    end
    
    return J*int(-E/2)
    
end;
```

```julia
function ΔE(S,i,j,N=length(S),L=int(sqrt(N)),J=1)
  
    return J*2*S[i,j]*(S[mod1(i+1,L),j]+S[mod1(i-1,L),j]+S[i,mod1(j-1,L)]+S[i,mod1(j+1,L)])
    
end;
```

```julia
function acceptance(S,T,i,j,N=length(S_old),L=int(sqrt(N)),J=1)
    
    Δe=ΔE(S,i,j,N,L,J)
    α=exp(-(1/T)*(Δe))
    
    if rand()<α
        return true,Δe
    else
        return false,0
    end
end;
```

```julia
function one_step_flip(S::Array,T,N=length(S),L=int(sqrt(N)),J=1)
    
    i,j=spin_choose(L)
    α,Δe=acceptance(S,T,i,j,N,L,J)
    if α==true
        spin_flip!(S,i,j,N,L)
        return Δe
    else
        return 0
    end
end;
```

```julia
function new_energy(E_old,Δe)
    
    return E_old+Δe
    
end;
```

```julia
function montecarlo_config_run(L::Int,steps::Int,T)
    N=L*L
    S=config_maker(L)

    for i in 1:steps
        one_step_flip(S,T,N,L,1)
    end
    
    return S
end;    
```

```julia
for T in [0.5:0.5:3]
    figure()
    S=montecarlo_config_run(10,10^5,T)
    imshow(S)
end
```

Podemos ver cómo, cuando va aumentando la temperatura, todos los espínes tienden a estár más desordenados, mientras que a temperaturas bajas, tienden a ser casi todos el mismo.

[4] Durante la simulación, rastrea los valores de la magnetización $M(t)$ y la energía $E(t)$. Dibuja estas cantidades como función del tiempo.

```julia
function montecarlo_energy_run(L::Int,steps::Int,T)
    N=L*L
    S=config_maker(L)
    E=[periodic_energy(S)]
    sizehint(E,steps)

    for i in 1:steps
        Δe=one_step_flip(S,T,N,L,1)
        push!(E,new_energy(E[i],Δe))
    end
    
    return E
end;    
```

```julia
magnetization(S)=sum(S);
```

```julia
function montecarlo_mag_run(L::Int,steps::Int,T)
    N=L*L
    S=config_maker(L)
    M=[magnetization(S)]
    sizehint(M,steps)

    for i in 1:steps
        one_step_flip(S,T,N,L,1)
        push!(M,magnetization(S))
    end
    
    return M
end;    
```

```julia
for T in [0.5:0.5:3]
    figure()
    E=montecarlo_energy_run(10,10^5,T)
    plot(E)
    title("Energía a temperatura $T con respecto al tiempo")
    xlabel("Tiempo")
    ylabel("Energía")
end
```

```julia
for T in [0.5:0.5:3]
    figure()
    M=montecarlo_mag_run(10,10^5,T)
    plot(abs(M))
    title("Magnetización a temperatura $T con respecto al tiempo")
    xlabel("Tiempo")
    ylabel("Magnetización")
end
```

[5] Quisiéramos calcular estimados de valores promedio como $\langle E \rangle_\beta$. Viendo tus gráficas, propón cómo hacer esto e impleméntalo. 

Vamos a calcular los valores promedio de las energías y de las magnetizaciones en función de la temperatura calculando el promedio de los arreglos de energía y de magnetizacipón obtenidos para una temperatura dada.

[6] Así, calcula $\langle E \rangle_\beta$ para distintos valores de $\beta$ entre $0$ y $5$, poniendo $J = 1$.

```julia
ET=Float64[]
T=[0.5:0.5:10]
sizehint(ET,length(T))
for temp in T
    E=montecarlo_energy_run(10,10^5,temp)
    push!(ET,mean(E))
end
plot(T,ET)
xlabel("Temperatura")
ylabel(L"Energia promedio $<E>_{T}$")
title("Energía promedio con respecto a la temperatura");
```

```julia
ET=Float64[]
T=[0.5:0.5:10]
sizehint(ET,length(T))
for temp in T
    E=montecarlo_energy_run(10,10^6,temp)
    push!(ET,mean(E))
end
plot([1/t for t in T],ET)
xlabel(L"$\beta$")
ylabel(L"Energia promedio $<E>_{\beta}$")
title(L"Energia promedio con respecto a $\beta$");
```
